### Prerequisites

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, binom
```

```python
X = 17
N = 116

PRIOR_ALPHA = 1
PRIOR_BETA = 4

POSTERIOR_ALPHA = 18
POSTERIOR_BETA = 103
```

### Task 1.4 

```python
theta = np.linspace(0, 1, 1000)
prior = beta(PRIOR_ALPHA, PRIOR_BETA)
posterior = beta(POSTERIOR_ALPHA, POSTERIOR_BETA)
likelihood = N * binom.pmf(X, N, theta)

prior_pdf = prior.pdf(theta)
posterior_pdf = posterior.pdf(theta)
```

```python
fig, ax1= plt.subplots(1, 1)
fig.set_size_inches(10, 8)
fig.suptitle("Prior, posterior, and normalized likelihood")
ax1.grid(linestyle='--')
ax1.set_xlabel('Theta, θ')
ax1.set_ylabel('Density')
ax1.plot(theta, prior_pdf, '-', label='Prior: Beta(1, 4)', c='r')
ax1.plot(theta, posterior_pdf, '-', label='Posterior: Beta(18, 103)', c='b')
ax1.plot(theta, likelihood, '-', label='Likelihood: Binomial(116, 17)', c='g')
ax1.legend()
plt.show() 
```

### Task 1.5

```python
posterior_prob_01 = posterior.cdf(0.1)
print(f"The posterior probability that θ < 0.1 is {posterior_prob_01:.4f}")
```

### Task 1.6

```python
interval_pc = 0.95
lower_limit_pc = (1.0 - interval_pc) / 2
upper_limit_pc = (1.0 + interval_pc) / 2

lower_limit_ci = posterior.ppf(lower_limit_pc)
upper_limit_ci = posterior.ppf(upper_limit_pc)

print(f"Central 95% posterior credible interval for θ is [{lower_limit_ci:.4f}, {upper_limit_ci:.4f}]")
```
